# Подготовка данных

```python
data = pd.read_csv("Customers.csv")
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data.columns
```

```python
sorted(set(data['Gender']))
```

```python
sorted(set(data['Profession']))
```

```python
gen_num = {gen: i for i, gen in enumerate(sorted(set(data['Gender'])))}
gen_num
```

```python
prof_num = {prof: i for i, prof in enumerate(sorted(set(data['Profession'])))}
prof_num
```

```python
try:
    data['Gender'] = data.Gender.apply(lambda gender: gen_num[gender])
except:
    pass
```

```python
try:
    data['Profession'] = data.Profession.apply(lambda prof: prof_num[prof])
except:
    pass
```

```python
try:
    data['Annual Income (K$)'] = data['Annual Income ($)']/1000
    data.drop(['Annual Income ($)'], axis=1, inplace=True)
except:
    pass
```

## Нормальность распределения

### Гистограммы рассеяния

Для определения нормальности распределения столбцов построим гистограммы рассеяния

```python
fig, ax = mpl.pyplot.subplots(
    nrows=1, ncols=data.shape[1], figsize=(30, 5)
)
colors = ["#555d8e","#566f94","#56819b","#5494a1","#62a4a7","#7db2ac","#97c0b0","#b1ceb5"]
for i, axi in enumerate(ax):
    axi.hist(
        data[data.columns[i]], 
        color = colors[i],
        edgecolor = "black"
    )
fig.legend(data.columns)
fig.show()
```

__Вывод__: Ни на одной гистограмме не наблюдается нормального распределения

### Статистический тест

Для определения нормальности распредения проведем статистический тест со следующей нулевой гипотезой: 
_Распределение столбцов является нормальным_

```python
for i, p in enumerate(sp.stats.normaltest(data).pvalue):
    print(f"p-критерий для столбца {data.columns[i]}: {p}\n")
```

p-критерий для каждого из столбцов стремится к нулю

__Вывод__: Нулевая гипотеза не выполняется, столбцы не имеют нормального распределения

## Нормализация

Нормализуем выбранные для кластеризации признаки

```python
from sklearn import preprocessing as prcss
data_scaled = pd.DataFrame(
    prcss.MinMaxScaler().fit_transform(data[data.columns[1::]]),
    columns = data.columns[1::]
)
data_scaled
```

Из полученных данных можно сделать вывод, что данные можно разделить на 2, 9 и 17 кластеров, используя признаки: Gender, Age, Spending Score (1-100), Profession, Work Experience, Family Size, Annual Income (K$). 2 по числу полов, 9 по числу профессий и 17 по опыту работы.

# Кластерный анализ

## Метод кластеризации K-means

Применим метод кластеризации K-means с предположенным ранее числом кластеров

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score as s_score
```

```python
kmeans = KMeans(17, init='k-means++', algorithm='elkan')
kmeans.fit(data_scaled)
```

```python
kmeans_gl = KMeans(17, init='k-means++', algorithm='elkan')
kmeans_gl.fit(data_scaled[1::])
```

Вычислим для него силуэт

```python
s_score(data_scaled, kmeans.predict(data_scaled))
```

```python
s_score(data_scaled[1::], kmeans_gl.predict(data_scaled[1::]))
```

Найдем количество кластеров, при котором значение силуэтта максимально

```python
clast = []
silhs = []
for i in np.arange(2, 18):
    kmeans = KMeans(i, init='k-means++', algorithm='lloyd')
    kmeans.fit(data_scaled)
    clast.append(i)
    silhs.append(s_score(data_scaled, kmeans.predict(data_scaled))) 
    print(f'Количество кластеров: {i}:', silhs[-1])
    
```

```python
fig, ax = mpl.pyplot.subplots(1, 1, figsize=(10,5))
ax.scatter(clast, silhs, color=colors[-1])
#ax.plot(clast, gaussian_filter1d(silhs, 1), color=colors[0])
ax.grid('minor')
ax.set_xlabel('Количество кластеров')
ax.set_ylabel('Размер силуэта')
#fig.legend(['Исходная', '], title='Зависимость силуэта\nот количества кластеров')
fig.show()
```

### Количество кластеров 2

```python
kmeans = KMeans(2, init='k-means++', algorithm='elkan')
kmeans.fit(data_scaled)
fig, ax = mpl.pyplot.subplots(1, 1, figsize=(10,5))
ax.scatter(data_scaled['Gender'], data_scaled['Annual Income (K$)'], c=kmeans.predict(data_scaled))
ax.set_xlabel('Пол')
ax.set_ylabel('Годовой доход')
```

### Количество кластеров 3

```python
kmeans = KMeans(3, init='k-means++', algorithm='elkan')
kmeans.fit(data_scaled)
fig, ax = mpl.pyplot.subplots(1, 1, figsize=(10,5))
ax.scatter(data_scaled['Profession'], data_scaled['Gender'], c=kmeans.predict(data_scaled))
ax.set_xlabel('Профессия')
ax.set_ylabel('Пол')
```
