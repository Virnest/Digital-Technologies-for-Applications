## 2.&nbsp;Линейные модели

```python
from sklearn import *
```

### 2.1.&nbsp;Простая линейная РЕГРЕССИЯ без оптимизации

```python
model_LR = linear_model.LinearRegression()
model_LR.fit(x_train, y_train)
```

```python
y_pred_LR = model_LR.predict(x_test)
```

Для использования в качестве классификатора надо вручную добавить бинаризацию по порогу. Для многоклассового разделения не подходит.

```python
from sklearn.preprocessing import Binarizer
```

```python
#LR - регроссор, возвращает значения из действительных чисел.
#Разделим их по порогу 0.5
transformer = Binarizer(threshold=0.5).fit(y_pred_LR.reshape(-1, 1))
y_pred_LR = transformer.transform(y_pred_LR.reshape(-1, 1))

print("Accuracy: "+str(metrics.accuracy_score(y_test, y_pred_LR)))
print("Precision: "+str(metrics.precision_score(y_test, y_pred_LR, average='macro')))
print("Recall: "+str(metrics.recall_score(y_test, y_pred_LR, average='macro')))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred_LR))
```

Для наглядности построим матрицу ошибок.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_LR)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.show()
```

### 2.3. Логистическая регрессия с подбором гиперпараметров

```python
#%pip install optuna
```

```python
import optuna as optuna
```

Описание экспериментов оптуны.

```python
def objective_LogReg(trial):
    params = {
        "random_state": 0,
        "max_iter": 1000,
        "C": trial.suggest_categorical(
            'C',
            [
                0.1,
                1,
                10,
                100,
                1000
            ]
        ),
        "solver": trial.suggest_categorical(
            "solver",
            [
                "lbfgs",
                "liblinear",
                "newton-cg",
                "newton-cholesky",
                "sag",
                "saga"
            ]
        )
    }
    # Создание необученной модели с гиперпараметрами, которые обозначены выше.
    model = linear_model.LogisticRegression(**params)
    model.fit(x_train,y_train) # Обучение модели на обучающей выборке
    predict = model.predict(x_test) # Оценка целевых значений для тестовой выборки

    # Расчёт метрики.
    # В данном случае используется recall - для правильно найденных женщин,
    # у которых присутсвует диабет.
    # average='binary' - считать по конкретному классу.
    # pos_label=1 - считать по классу с меткой 1, т.е. только по больным.
    return metrics.recall_score(y_test, predict, average='binary', pos_label=1)
```

```python
# direction="maximize" - хотим подобрать максимально возможное значение метрики
# recall - максимальную долю найденных женщин с диабетом.
study = optuna.create_study(direction="maximize")
# Отключение логирования
# optuna.logging.set_verbosity(optuna.logging.WARNING)
study.optimize(objective_LogReg, n_trials=100)
```

```python
study.best_params
```

```python
model_LogReg_optuned = linear_model.LogisticRegression(
    random_state= 0,
    max_iter = 1000,
    solver = "sag",
    C = 100
)
model_LogReg_optuned.fit(x_train,y_train)
y_pred_LogReg_optuned = model_LogReg_optuned.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_LogReg_optuned))
```

### 5.1. Лес случайных деревьев
https://scikit-learn.ru/1-11-ensemble-methods/#forests-of-randomized-trees

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# n_estimators=15 - количество деревьев в лесу.
# max_depth=5, - ограничиваем глубину всех деревьев.
model_rf = RandomForestClassifier(
    n_estimators=20,
    max_depth=8,
    random_state=1
)
model_rf.fit(x_train,y_train)
```

```python
y_pred_rf = model_rf.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
def objective_RF(trial):
    params = {
        "random_state": 0,
        "n_estimators": trial.suggest_int("n_estimators", 4, 100),
        "max_depth": trial.suggest_int("max_depth", 4, 30),
    }
    # Создание необученной модели с гиперпараметрами, которые обозначены выше.
    model = RandomForestClassifier(**params)
    model.fit(x_train,y_train) # Обучение модели на обучающей выборке
    predict = model.predict(x_test) # Оценка целевых значений для тестовой выборки

    # Расчёт метрики.
    # В данном случае используется recall - для правильно найденных женщин,
    # у которых присутсвует диабет.
    # average='binary' - считать по конкретному классу.
    # pos_label=1 - считать по классу с меткой 1, т.е. только по больным.
    return metrics.recall_score(y_test, predict, average='binary', pos_label=1)
```

```python
# direction="maximize" - хотим подобрать максимально возможное значение метрики
# recall - максимальную долю найденных женщин с диабетом.
study = optuna.create_study(direction="maximize")
# Отключение логирования
optuna.logging.set_verbosity(optuna.logging.INFO)
study.optimize(objective_RF, n_trials=100)
```

```python
study.best_params
```

```python
# n_estimators=15 - количество деревьев в лесу.
# max_depth=5, - ограничиваем глубину всех деревьев.
model_rf1 = RandomForestClassifier(
    n_estimators=22,
    max_depth=9,
    random_state=1
)
model_rf1.fit(x_train,y_train)
y_pred_rf1 = model_rf1.predict(x_test)
print(classification_report(y_test, y_pred_rf1))
```

## 6.&nbsp;Методы автоматизации перебора гиперпараметров

Позволяют автоматически перебрать значения гиперпараметров модели в заданных диапазонах с фиксированными шагами или с градиентным спуском.

### 6.1. Optuna
https://optuna.org/

```python
#!pip install optuna
```

```python
import optuna as optuna
```

```python
def objective(trial):
    params = {
        "loss_function": "Logloss",
        'iterations': 500,
        'l2_leaf_reg': trial.suggest_int('l2_leaf_reg',0,10,step=1),
        'learning_rate': trial.suggest_float('learning_rate',0.01, 0.5, step=0.01),
        'rsm':  trial.suggest_float('rsm',0.1, 0.7, step=0.1),

        #"objective": trial.suggest_categorical("objective", ["RMSE", "MAPE"]),
        #"colsample_bylevel": trial.suggest_float("colsample_bylevel", 0.01, 0.1, log=True),
        "depth": trial.suggest_int("depth", 4, 8),
        "boosting_type": trial.suggest_categorical("boosting_type", ["Ordered", "Plain"]),
        "bootstrap_type": trial.suggest_categorical(
            "bootstrap_type", ["Bayesian", "Bernoulli", "MVS"]
        ),
        #"used_ram_limit": "3gb",
        #"eval_metric": "Accuracy",
    }
    model = CatBoostClassifier(**params, verbose=False)
    model.fit(x_train,y_train)
    predict = model.predict(x_test)
    return metrics.precision_score(y_test, predict, average='macro')
```

```python
#Работает долго!
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=20)
```

```python
study.best_trial
```

```python
study.best_params
```

```python
fig = optuna.visualization.plot_optimization_history(study)
fig.show()
```

```python
fig = optuna.visualization.plot_param_importances(study)
fig.show()
```

```python
fig = optuna.visualization.plot_contour(study, params=["learning_rate", "rsm"])
fig.show()
```

```python
model_cb_optuned = CatBoostClassifier(**study.best_params,verbose = False)
model_cb_optuned.fit(x_train,y_train)
y_pred_cb_optuned = model_cb_optuned.predict(x_test)
print("Accuracy: "+str(metrics.accuracy_score(y_test, y_pred_cb_optuned)))
print("Precision: "+str(metrics.precision_score(y_test, y_pred_cb_optuned, average='macro')))
print("Recall: "+str(metrics.recall_score(y_test, y_pred_cb_optuned, average='macro')))
```
