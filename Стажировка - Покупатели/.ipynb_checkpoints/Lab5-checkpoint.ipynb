## 2.&nbsp;Вариант с ошибкой (для демонстрации, не повторять)

### 2.1.&nbsp;Оценка количества кластеров для алгоритма KMeans


Сначала общим циклом расчёт всех метрик, затем, ниже, подробнее по каждой метрике.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
def calculate_metrics(df, max_clusters):
    km_scores= []
    silhouette_scores = []
    db_scores = []

    for i in range(2, max_clusters):
        print("Количество кластеров: {}".format(i))

        km = KMeans(n_clusters=i, random_state=0, n_init=1).fit(df)
        preds = km.predict(df)

        km_score = -km.score(df)
        km_scores.append(km_score)
        print("Метрика K-Means (локоть): {}".format(km_score))

        silhouette = silhouette_score(df,preds)
        silhouette_scores.append(silhouette)
        print("Метрика силуэтов: {}".format(silhouette))

        db = davies_bouldin_score(df,preds)
        db_scores.append(db)
        print("Метрика Davies Bouldin: {}".format(db))

        print("-"*100)
    return {"km": km_scores, "silhouette":silhouette_scores, "db": db_scores}
```

```python
#Максимальное количество кластеров для рассмотрения.
# 1 и более 20 не имеет смысла рассматривать.
max_clusters = 11
```

```python
scores = calculate_metrics(X, max_clusters)
```

```python
def print_scores(scores):
    x=[i for i in range(2,2+len(scores["km"]))]

    plt.figure(figsize=(7,4))
    plt.title("Метод локтя для определения количества кластеров\n",fontsize=16)
    plt.scatter(x = x, y=scores["km"], s=150, edgecolor='k')
    plt.grid(True)
    plt.xlabel("Количество кластеров, шт",fontsize=14)
    plt.ylabel("Метрика K-Means, попугаев",fontsize=15)
    plt.xticks(x,fontsize=14)
    plt.yticks(fontsize=14)
    plt.show()

    plt.figure(figsize=(7,4))
    plt.title("Метрика силуэтов для определения количества кластеров\n",fontsize=16)
    plt.scatter(x=x,y=scores["silhouette"],s=150,edgecolor='k')
    plt.grid(True)
    plt.xlabel("Количество кластров, шт",fontsize=14)
    plt.ylabel("Метрика силуэтов, попугаев",fontsize=15)
    plt.xticks(x,fontsize=14)
    plt.yticks(fontsize=14)
    plt.show()

    plt.scatter(x=x,y=scores["db"],s=150,edgecolor='k')
    plt.grid(True)
    plt.xlabel("Количество кластров, шт",fontsize=14)
    plt.ylabel("Метрика Davies-Bouldin, попугаев",fontsize=14)
    plt.xticks(x,fontsize=14)
    plt.yticks(fontsize=14)
    plt.show()
```

#### 2.1.1. Метрика алгоритма KMeans (метод локтя)

Метрикой алгоритма KMeans является сумма квадратов ошибок внутри кластеров (SSE, J-score или WCSS) - расстояний от элемента до центра кластера. Возвращается методом km.score(X) в отрицательном варианте. Отрицательность, наверно, добавили для удобства оптимизации.

```python
Math(r"\sum_{i=0}^{n}\min_{\mu_j \in C}(||x_i - \mu_j||^2)")
```

На графике (рисунок 1) нужно представить, что линия - это рука. Надо найти количество кластеров, при котором у руки находится "изгиб локтя" (максимальный перегиб).

#### 2.1.2. Метод силуэтов.

Метрика рассчитывается по формуле (b - a) / max(a, b) для каждого элемента множества, где a - среднее расстояние от текущей точки до всех точек своего кластера, b - расстояние от текущей точки до центра ближайщего соседнего кластера. Метод silhouette_score() возвращает среднее значение по всем точкам множества.

На графике (рисунок 2) нужно выбрать точку(и), которые сильнее всего выбиваются из общей последовательности ВНИЗ (меньше-лучше).

#### 2.1.3. Метрика [Davies-Bouldin](https://en.wikipedia.org/wiki/Davies–Bouldin_index)

Формулы приведены в википедии по ссылке выше.

На графике (рисунок 3) нужно выбрать точку(и), которые сильнее всего выбиваются из общей последовательности ВНИЗ (меньше-лучше).

```python
print_scores(scores)
```

### 2.2.&nbsp;Кластеризация методом [К-средних](https://ru.wikipedia.org/wiki/Метод_k-средних) (K-Means)

image.png

Расстояние между точками и центрами определяется, чаще всего, по формуле Евклида.

**Проблемы:**
* Не гарантируется достижение глобального минимума суммарного квадратичного отклонения V, а только одного из локальных минимумов.
* Результат зависит от выбора исходных центров кластеров, их оптимальный выбор неизвестен.
* Число кластеров надо знать заранее (оценивали в предыдущем пункте).

```python
model_cluster_6 = KMeans(6, n_init=10)
model_cluster_6.fit(X)
y_pred_km_6 = model_cluster_6.predict(X)
```

```python
model_cluster_13 = KMeans(13, n_init=10)
model_cluster_13.fit(X)
y_pred_km_13 = model_cluster_13.predict(X)
```

```python
x= X['sqft_living']
y = X['price']
fig, ax = plt.subplots(figsize=(20,10))
sc = ax.scatter(x, y, c=y_pred_km_6, cmap='tab20')
ax.grid(True)
plt.xlabel("Жилая площадь, кв.фут.",fontsize=14)
plt.ylabel("Стоимость, $",fontsize=15)
ax.legend(*sc.legend_elements(), title='Кластеры')
```

## 3.&nbsp;Корректный вариант (можно использовать для самостоятельной работы)

### 3.1.&nbsp;Дополнительная предобработка данных

Чтобы избежать эффекта, что признак с большими по модулю значениями вносит больший вклад при расчёте расстояния (кластеры располагаются полосками на графиках), нужно привести все значения всех признаков к единому масштабу, например, [0, 1].

```python
#Здесь фактическое шкалирование
from sklearn.preprocessing import MinMaxScaler
scalers = {}
dfs = X.copy(deep=True) #Перед изменением полностью копируем данные.
for c in X.columns:
    scalers[c] = MinMaxScaler().fit(dfs[c].values.reshape(-1,1))
    dfs[c] = scalers[c].transform(dfs[c].values.reshape(-1,1))
```

```python
dfs
```

### 3.2.&nbsp;Оценка количества кластеров для алгоритма KMeans

```python
scores = calculate_metrics(dfs, max_clusters)
print_scores(scores)
```

### 3.3.&nbsp;Кластеризация методом KMeans

```python
model_km_s = KMeans(4, n_init=10)
model_km_s.fit(dfs)
y_pred_km_s = model_km_s.predict(dfs)
```

```python
x= dfs['sqft_living']
y = dfs['price']
fig, ax = plt.subplots(figsize=(20,10))
sc = ax.scatter(x, y, c=y_pred_km_s, cmap='tab20')
ax.grid(True)
plt.xlabel("Жилая площадь, доли от макисмальной",fontsize=14)
plt.ylabel("Стоимость, доли от макисмальной",fontsize=15)
ax.legend(*sc.legend_elements(), title='Кластеры')
```

Почему все кластеры на графике всё равно перемешены друг с другом?

### 3.4.&nbsp;Кластеризация методом DBSCAN

```python
model_dbscan_s = DBSCAN(eps=0.1, min_samples=100)
y_pred_dbscan_s = model_dbscan_s.fit_predict(dfs)
```

```python
model_dbscan_s.get_params()
```

```python
x= dfs['sqft_living']
y = dfs['price']
fig, ax = plt.subplots(figsize=(20,10))
sc = ax.scatter(x, y, c=y_pred_dbscan_s, cmap='tab20')
ax.grid(True)
plt.xlabel("Жилая площадь, доли от макисмальной",fontsize=14)
plt.ylabel("Стоимость, доли от макисмальной",fontsize=15)
ax.legend(*sc.legend_elements(), title='Кластеры')
```

Для проверки выведем список номеров кластеров. Как видно, номер -1 тоже присутствует.

```python
np.unique(y_pred_dbscan_s)
```

### 3.5.&nbsp;Анализ результатов кластеризации

Обратное преобразование к обычному масштабу координат центров кластеров.

```python
centers = []
for i in range(0,len(model_km_s.cluster_centers_)):
  coordinates = {}
  for j in range(0,len(X.columns)):
    coordinates[X.columns[j]] = scalers[X.columns[j]].inverse_transform(model_km_s.cluster_centers_[i][j].reshape(-1, 1))[0][0]
  centers.append(coordinates)
centers = pandas.DataFrame(centers)
centers
```

## 4.&nbsp; Уменьшение размерности

[Стохастическое вложение соседей с t-распределением ](https://ru.wikipedia.org/wiki/Стохастическое_вложение_соседей_с_t-распределением) (t-SNE)

```python
from sklearn.manifold import TSNE
```

```python
X.shape
```

```python
# Работает долго!
X_embedded = TSNE(
    n_components=2,
    learning_rate='auto',
    init='random',
    perplexity=3
).fit_transform(X)
X_embedded.shape
```

Из 13-мерного пространства Х удалось перейти к 2-мерному синтетическому пространству X_embedded, которое можно наглядно изобразить на плоском графике.

X_embedded объект типа ndarray, a не dataframe.

```python
X_embedded[0][0]
```

```python
X_embedded[:,0]
```

Проведение кластеризации в новой, синтетической плоскости.

```python
model_km_embedded = KMeans(4, n_init=1)
model_km_embedded.fit(X_embedded)
y_pred_km_embedded = model_km_embedded.predict(X_embedded)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sc = ax.scatter(X_embedded[:,0], X_embedded[:,1], c=y_pred_km_embedded, cmap='tab10')
ax.grid(True)
ax.set_xlabel("Синтетический признак 1, попугаи",fontsize=14)
ax.set_ylabel("Синтетический признак 2, попугаи",fontsize=15)
ax.legend(*sc.legend_elements(), title='Кластеры')
```

К сожалению, реализация метода t-SNE в библиотеке scikit-learn не подразумевает обратного преобразования. Принадлежность элементов множества к кластерам сохраняется по порядковым номерам. Центры кластеров в исходном многомерном пространстве можно вычислить по формуле координат цетра масс, с использованием данных из исходного множества и данных об их принадлежности к кластерам.

Также можно попробовать применить другие методы:
* PCA для плотных/концентрированных данных;
* TruncatedSVD для разреженных данных.
