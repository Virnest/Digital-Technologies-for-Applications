```python
from sklearn.cluster import DBSCAN, KMeans
```

```python
df_s = (df_merg - df_merg.min())/(df_merg.max() - df_merg.min())
```

```python
criterion = []
# Количество кластеров будем изменять от 2 до 7
for k in range(2,10):
    kmeans = KMeans(n_clusters = k, init = 'k-means++', max_iter = 2000, n_init = 10, random_state = 42)
    kmeans.fit(df_merg)
    criterion.append(kmeans.inertia_)

import matplotlib.pyplot as plt
plt.plot(range(2,10), criterion)
```

```python
x = df_s
```
