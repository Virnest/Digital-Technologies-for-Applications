```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("diabetes.csv")
data
```

```python
data.info()
```

```python
data.value_counts()
```

```python
X = data[data.columns[:-1:]]
y = data[data.columns[-1]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
```

# SVM - SVC

```python
svmReg = svm.SVC()
```

```python
svmReg.fit(X_train, y_train)
```

```python
plt.plot(svmReg.predict(X_test), '.')
plt.plot(y_test, '.')
```

```python
print(classification_report(y_test.astype(float), svmReg.predict(X_test)))
```

```python
cross_val_score(svmReg, X=X_train, y=y_train)
```

```python
X_train_scl = (X_train - X_train.min())/(X_train.max() - X_train.min())
X_test_scl = (X_test - X_test.min())/(X_test.max() - X_test.min())
cross_val_score(svmReg, X=X_train_scl, y=y_train)
```

```python
svmReg.fit(X_train_scl, y_train)
```

```python
plt.plot(svmReg.predict(X_test_scl), '.')
plt.plot(y_test, '.')
```

```python
for kernel in ['linear', 'poly', 'rbf', 'sigmoid']:
    svmReg = svm.SVC(kernel=kernel)
    svmReg.fit(X_train_scl, y_train)
    plt.plot(y_test, '.')
    plt.plot(svmReg.predict(X_test_scl), '.', label=f'{kernel}')
    plt.legend()
    plt.show()
```

# LogisticRegression

```python
linReg = LogisticRegression(max_iter = 1000)
```

```python
linReg.fit(X_train, y_train)
```

```python
plt.plot(linReg.predict(X_test), '.')
plt.plot(y_test, '.')
```

```python
print(classification_report(y_test.astype(float), linReg.predict(X_test)))
```

```python
cross_val_score(linReg, X=X_train, y=y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
GridSearchCV(linReg, param_grid=[])
```

# KNeighborsClssifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knnn = KNeighborsClassifier()
```

```python
knnn.fit(X_train, y_train)
```

```python
plt.plot(knnn.predict(X_test),'.')
plt.plot(y_test, '.')
```

```python
print(classification_report(y_test, knnn.predict(X_test)))
```
